```python
from utils.utils import *
from PIL import Image
from tensorflow import stack, concat, expand_dims, reshape, transpose
```

```python
im = Image.open('butterfly.jpeg').convert("L")
w, h = im.size   # Get dimensions

left   = (w - 191) // 2
top    = (h - 191) // 2
right  = left + 191
bottom = top + 191

# Crop the center of the image
im = np.array(im.crop((left, top, right, bottom)).resize((256, 256)), dtype=np.float32 )
plt.imshow(im)

```

```python

s, a, f = getSymAntiSymTF(reshape(im, (256,256,1,1)))
```

```python
s.shape
```

```python
vmin = min(np.array(s).min(), np.array(a).min())
vmax = max(np.array(s).max(), np.array(a).max())
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8,4))

axs[0].imshow(s[...,0,0], cmap="gray", vmin=vmin, vmax=vmax)
axs[0].set_title("Image 1")
axs[0].axis("off")

axs[1].imshow(a[...,0,0], cmap="gray", vmin=vmin, vmax=vmax)
axs[1].set_title("Image 2")
axs[1].axis("off")

plt.tight_layout()
plt.show()
```
